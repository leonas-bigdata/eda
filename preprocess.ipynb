# Data Preprocessing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from collections import Counter
import warnings
warnings.filterwarnings("ignore")
```

```python
def load_data(data_file):
    users, items = [], []
    interactions_per_user = {}

    with open (data_file) as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) < 2:
                continue
            user_id = int(parts[0])
            item_ids = [int(x) for x in parts[1:]]
            interactions_per_user[user_id] = item_ids
            users.extend([user_id] * len(item_ids))
            items.extend(item_ids)
    return np.array(users), np.array(items), interactions_per_user

def analyse_dataset(current_dataset):
    train_users, train_items, train_dict = load_data(f"{current_dataset}/train.txt")
    test_users, test_items, test_dict = load_data(f"{current_dataset}/test.txt")

    user_list_path = f"{current_dataset}/user_list.txt"
    item_list_path = f"{current_dataset}/item_list.txt"
    
    if user_list_path:
        user_list = pd.read_csv(f"{current_dataset}/user_list.txt", sep =" ")
    if item_list_path:
        item_list = pd.read_csv(f"{current_dataset}/item_list.txt", sep =" ")
    

    n_users = len(user_list)
    n_items = len(item_list)
    n_train = len(train_users)
    n_test = len(test_users)
    train_users_set = set(train_users)
    test_users_set  = set(test_users)
    train_items_set = set(train_items)
    test_items_set  = set(test_items)

    density_train = n_train / (n_users * n_items)
    density_test  = n_test  / (n_users * n_items)

    print(f"{current_dataset} dataset overview")
    print(f"Total users (user_list)   : {n_users:,}")
    print(f"Total items (item_list)   : {n_items:,}")
    print(f"Train interactions        : {n_train:,}")
    print(f"Test  interactions        : {n_test:,}")
    print(f"Train+Test interactions   : {n_train + n_test:,}")
    print(f"Train sparsity            : {density_train:.6f}  ({density_train*100:.6f}%)")
    print(f"Test  sparsity            : {density_test:.6f}  ({density_test*100:.6f}%)")
    
    print("\nUser Coverage")
    print(f"Users in train            : {len(train_users_set):,} / {n_users:,} ({len(train_users_set)/n_users*100:.3f}%)")
    print(f"Users in test             : {len(test_users_set):,} / {n_users:,} ({len(test_users_set)/n_users*100:.3f}%)")

    
    print("\nItem Coverage")
    print(f"Items in train: {len(train_items_set):,} / {n_items:,} ({len(train_items_set)/n_items*100:.3f}%)")
    print(f"Items in test: {len(test_items_set):,} / {n_items:,} ({len(test_items_set)/n_items*100:.3f}%)")
    
    train_counts = np.array([len(v) for v in train_dict.values()])
    test_counts  = np.array([len(v) for v in test_dict.values()])
    print("\nInteractions per User Train and Test")
    print(f"  Mean : {train_counts.mean():.2f}, {test_counts.mean():.2f}")
    print(f"  Median  : {np.median(train_counts):.1f}, {np.median(test_counts):.1f}")
    print(f"  Std     : {train_counts.std():.2f}, {test_counts.std():.2f}")
    print(f"  Min     : {train_counts.min()}, {test_counts.min()}")
    print(f"  Max     : {train_counts.max()}, {test_counts.max()}")


    #4. Popularity
    train_item_counts = Counter(train_items)
    test_item_counts  = Counter(test_items)
    train_pop = np.array(list(train_item_counts.values()))
    test_pop  = np.array(list(test_item_counts.values()))
    
    print(f"Mean interactions per item  : {train_pop.mean():.5f}")
    print(f"Median : {np.median(train_pop):.5f}")
    print(f"Max (most popular) : {train_pop.max()}")
    topk_items = train_item_counts.most_common(20)
    print(f"  Top-20 items (id, count) : {topk_items}")

```

```python
"""
Script run to fix amazondataset literal issue
Bug fixed: ValueError: invalid literal for int() with base 10: ''
"""
import os
def clean_file(input_path, choice = 1):
    """Clean train.txt / test.txt of amazon-book"""
    output_path = input_path  

    lines_kept = 0
    lines_skipped = 0
    cleaned_lines = []

    with open(input_path, "r", encoding="utf-8") as f:
        for line_num, raw in enumerate(f, 1):
            line = raw.strip()

            if not line:
                lines_skipped += 1
                continue
            tokens = [t for t in line.split(" ") if t != ""]
            if len(tokens) < 2:
                print(f"[SKIP] line {line_num}: only has {len(tokens)} token")
                lines_skipped += 1
                continue

            try:
                [int(t) for t in tokens]
            except ValueError as e:
                print(f"[SKIP] line {line_num}: {e}")
                lines_skipped += 1
                continue

            cleaned_lines.append(" ".join(tokens))
            lines_kept += 1

    if choice == 2:
        print('Delete four single token lines')
        with open(output_path, "w", encoding="utf-8") as f:
            f.write("\n".join(cleaned_lines) + "\n")

    print(f"{input_path}: kept={lines_kept}, skipped={lines_skipped}")


amazon_dir = "data/amazon-book"   

for filename in [ "test.txt", "train.txt" ]:
    path = os.path.join(amazon_dir, filename)
    if os.path.exists(path):
        clean_file(path, 2)
    else:
        print(f"Cannot found: {path}")
```

**Movie Len**

For movielen dataset, the data format is a bit different

The format is a list containing lists, for each item of that list is the item id that the user_id at the current index are interacted with



```python
def format_movielen(dataset_name):
    import json
    if "movielen" not in dataset_name:
        print('Data set should be movielen')
        return
    output_path = dataset_name.replace(".json",".txt")
    output_path = output_path.replace("_data","")
    with open(dataset_name, "r") as f:
        data = json.load(f)
    with open(output_path, 'w') as f:
        for user_id, items in enumerate(data):
            if items:
                line = str(user_id) + ' ' + ' '.join(str(i) for i in items)
                f.write(line + '\n')
    print(f'Conversion {dataset_name} complete')

def analyse_dataset_for_movielen(current_dataset):
    train_users, train_items, train_dict = load_data(f"{current_dataset}/train.txt")
    test_users, test_items, test_dict = load_data(f"{current_dataset}/test.txt")

    # Derive n_users and n_items from union of train + test
    all_users = set(train_users) | set(test_users)
    all_items = set(train_items) | set(test_items)
    n_users = len(all_users)
    n_items = len(all_items)
    n_train = len(train_users)
    n_test  = len(test_users)

    train_users_set = set(train_users)
    test_users_set  = set(test_users)
    train_items_set = set(train_items)
    test_items_set  = set(test_items)

    density_train = n_train / (n_users * n_items)
    density_test  = n_test  / (n_users * n_items)

    print(f"{current_dataset} dataset overview")
    print(f"Total users (union)       : {n_users:,}")
    print(f"Total items (union)       : {n_items:,}")
    print(f"Train interactions        : {n_train:,}")
    print(f"Test  interactions        : {n_test:,}")
    print(f"Train+Test interactions   : {n_train + n_test:,}")
    print(f"Train sparsity            : {density_train:.6f}  ({density_train*100:.6f}%)")
    print(f"Test  sparsity            : {density_test:.6f}  ({density_test*100:.6f}%)")

    print("\nUser Coverage")
    print(f"Users in train            : {len(train_users_set):,} / {n_users:,} ({len(train_users_set)/n_users*100:.3f}%)")
    print(f"Users in test             : {len(test_users_set):,} / {n_users:,} ({len(test_users_set)/n_users*100:.3f}%)")

    print("\nItem Coverage")
    print(f"Items in train            : {len(train_items_set):,} / {n_items:,} ({len(train_items_set)/n_items*100:.3f}%)")
    print(f"Items in test             : {len(test_items_set):,} / {n_items:,} ({len(test_items_set)/n_items*100:.3f}%)")

    train_counts = np.array([len(v) for v in train_dict.values()])
    test_counts  = np.array([len(v) for v in test_dict.values()])
    print("\nInteractions per User (Train / Test)")
    print(f"  Mean   : {train_counts.mean():.2f}  /  {test_counts.mean():.2f}")
    print(f"  Median : {np.median(train_counts):.1f}  /  {np.median(test_counts):.1f}")
    print(f"  Std    : {train_counts.std():.2f}  /  {test_counts.std():.2f}")
    print(f"  Min    : {train_counts.min()}  /  {test_counts.min()}")
    print(f"  Max    : {train_counts.max()}  /  {test_counts.max()}")

    train_item_counts = Counter(train_items)
    test_item_counts  = Counter(test_items)
    train_pop = np.array(list(train_item_counts.values()))
    test_pop  = np.array(list(test_item_counts.values()))

    print(f"\nItem Popularity (Train)")
    print(f"  Mean   : {train_pop.mean():.5f}")
    print(f"  Median : {np.median(train_pop):.5f}")
    print(f"  Max    : {train_pop.max()}")
    print(f"  Top-20 items (id, count) : {train_item_counts.most_common(20)}")

format_movielen("data/movielen/train_data.json")
format_movielen("data/movielen/test_data.json")
```

```python
analyse_dataset_for_movielen("data/movielen")
```
